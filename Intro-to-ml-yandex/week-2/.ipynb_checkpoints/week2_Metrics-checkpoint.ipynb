```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
sc = []
for p in np.linspace(start=1, stop=10, num=200):
    scores = []
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        model = KNeighborsRegressor(
            n_neighbors=5, 
            n_jobs=-1, 
            weights='distance', 
#             scoring='neg_mean_squared_error',
            metric=(lambda x, y: sum(abs(x - y)**p)**(1/p)),
#             minkowski_distance=p
        )
        model.fit(X_train, y_train)
        scores.append(model.score(X_test, y_test))
    print(p, ":", st.mean(scores))
    sc.append(st.mean(scores))
print("Best:", np.argmax(sc)+1, np.max(sc))
plt.plot(np.linspace(start=1, stop=10, num=200), sc);
```
