```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data-logistic.csv", header=None)
```

```python
plt.scatter(df[[1]], df[[2]], c=df[[0]]);
```

```python
def init(lr = 0.001, c = 10, num_iter = 10, L=2):
    n = df.shape[0] # num of examples
    f = 2 # num of features
    x = df.iloc[:, 1:].as_matrix()    # (205, 2)
    w = np.zeros((1, f))                   # (2, )
    y = df.iloc[:, 0].as_matrix()     # (205, 1)
    y = np.reshape(y, (n, 1))
    return lr, c, n, f, x, w, y, num_iter, L

lr, c, n, f, x, w, y, num_iter, L = init()
print("x:", x.shape)
print("w:", w.shape)
print("y:", y.shape)
```

```python
def sigmoid(x, w):
    # used to calculate loss and gradients
    sg = 1./(1.+np.exp(-y*np.dot(x, w.T))) # (205, 1)
#     print(sg)
    return sg

def sigmoid_v2(x, w):
    # used to get predictions
    sg = 1./(1.+np.exp(-np.dot(x, w.T))) # (205, 1)
#     print(sg)
    return sg
```

```python
def loss(w):
    reg = 1./2.*c*np.linalg.norm(w, ord=L)
    loss = 1./n*np.sum(np.log(1./sigmoid(x, w))) + reg
    return loss
```

```python
def update_grad(w):
    reg = lr*c*w
    if L==1:
        reg = reg*2.
#     w = w + lr/n * np.dot((sigmoid(x, w) - y).T, x) - reg
    w = w + lr/n * np.dot((y*(-sigmoid(x, w) + 1.)).T, x) - reg
    return w
```

```python
lr, c, n, f, x, w, y, num_iter, L = init(c=0.4, lr=0.001, num_iter=10000, L=2)
losses = [loss(w)]
weights = [w[0]]
i = 0
while i <= num_iter:
    i = i + 1
    new_w = update_grad(w)
    losses.append(loss(new_w))
    weights.append(new_w[0])
    if np.linalg.norm(w-new_w, ord=2) <= 10e-7:
        w = new_w
        break
    w = new_w
    
print("initial loss:", losses[0])
print("loss:", losses[-1])
print("roc_auc_score:", round(roc_auc_score(y, sigmoid_v2(x, w)), 3))
print("weights:", weights[-1])

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_figwidth(15)

ax1.plot(losses)
ax1.set_title("loss")

ax2.plot([wt[0] for wt in weights])
ax2.plot([wt[1] for wt in weights])
ax2.set_title("weights")
```
